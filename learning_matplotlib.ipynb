## Functional Plot

```python
x1 = np.linspace(0,5,10)
y1 = x1 ** 2
plt.plot(x1, y1)
plt.title('Function')
plt.xlabel('Days')
plt.ylabel('Sqaures')
```

## Multiple Plots

```python
plt.subplot(1, 2, 1)
plt.plot(x1, y1, 'r')
plt.subplot(1, 2, 2)
plt.plot(x1, y1, 'b')
```

## Figure object

```python
fig1 =  plt.figure(figsize=(5,4), dpi=100)
ax1 = fig1.add_axes([0.1,0.1,0.9,0.9])
ax1.set_xlabel('Days')
ax1.set_ylabel('Days Squared')
ax1.set_title('Days Sqaured Chart')
ax1.plot(x1, y1, label='x/x2')
ax1.plot(y1, x1, label='x2/x')
# upper right: 1, upper left: 2, lower left 3, lower right: 4, or supply a tuple of x,y from lower left
ax1.legend(loc=1)

ax2 = fig1.add_axes([0.45,0.45,0.4,0.3])
ax2.set_xlabel('Days')
ax2.set_ylabel('Days Squared')
ax2.set_title('Days Sqaured Chart')
ax2.plot(x1, y1, 'r')
ax2.text(0,40, 'Message')
```

## Subplot

```python
fig2, ax2 = plt.subplots(figsize=(16,8), nrows=1, ncols=3)
plt.tight_layout()
ax2[1].set_title('Plot 2')
ax2[1].set_xlabel('X')
ax2[1].set_ylabel('X Sqaured')
ax2[1].plot(x1, y1)
```

## Appearance options

```python
# Default colors (b: blue, g: green, r: red, c: cyan, m: mgenta, y: yellow, k: black, w: white)
# color='0.75' creates a 75% gray
# hexcodes: color='#eeefff'
# color names: color='burlywood' # https://en.wikipedia.org/wiki/Web_colors
# line styles: https://matplotlib.org/gallery/lines_bars_and_markers/line_styles_reference.html
# markers: https://matplotlib.org/3.3.3/api/markers_api.html

fig3 = plt.figure(figsize=(6,4))
ax3 = fig3.add_axes([0,0,1,1])
ax3.plot(x1, y1, color='navy', alpha=.75, lw=2, ls='-.', marker='o', markersize=10, markerfacecolor='r', markeredgecolor='y', markeredgewidth=4)
ax3.set_xlim([0,3])
ax3.set_ylim([0,8])

ax3.set_title('Plot 3')
ax3.set_xlabel('X')
ax3.set_ylabel('X Sqaured')

# dashes: 5pt line, 2pt break, 1pt line 2pt break
ax3.grid(True, color='0.6', dashes=(5,2,1,2))
ax3.set_facecolor('#faebd7')
```

## Pie Chart

```python
import random

fig6 = plt.figure(figsize=(8,5), dpi=100)
ax6 = fig6.add_axes([0.1,0.1,.9,.9])

types = ['Water', 'Normal', 'Flying', 'Grass', 'Psychic', 'Bug', 'Fire', 'Poison', 'Ground', 'Rock',
         'Fighting', 'Dark', 'Steel', 'Electric', 'Dragon', 'Fairy', 'Ghost', 'Ice']
poke_num = [133, 109, 101, 98, 85, 77, 68, 66, 65, 60, 57, 54, 53, 51, 50, 50, 46, 40]

colors = []
for i in range(len(types)):
    rgb = (random.uniform(0, .5), random.uniform(0, .5), random.uniform(0, .5))
    colors.append(rgb)

explode = [0] * 18
explode[0] = 0.2

wedges, texts, autotexts = plt.pie(poke_num, explode=explode, labels=types, colors=colors,
                                   autopct='%1.0f%%', shadow=True, startangle=140, textprops=dict(color='w'))
plt.legend(wedges, types, loc='right', bbox_to_anchor=(1, 0, 0.5, 1))
```

## Plot timeseries

```python
from datetime import datetime
data = pd.read_csv('GOOG.csv', index_col='Date')['2020-05-20':'2020-08-19']
print(data)
arr = data.to_numpy()
cp = arr[:,4]
holidays = [datetime(2020,5,25), datetime(2020,8,19)]

# freq define: https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#timeseries-offset-aliases
date_arr = pd.bdate_range(start='5/20/2020', end='8/19/2020', freq='C', holidays=holidays)
arr1 = date_arr.to_numpy()
fig7 = plt.figure(figsize=(8,5))
ax7 = fig7.add_axes([0.1,0.1,0.9,0.9])
plt.plot(arr1, cp)
```

## Tables

```python
data['Open'] = pd.Series([round(val, 2) for val in data['Open']], index=data.index)
data['Close'] = pd.Series([round(val, 2) for val in data['Close']], index=data.index)
data['High'] = pd.Series([round(val, 2) for val in data['High']], index=data.index)
data['Low'] = pd.Series([round(val, 2) for val in data['Low']], index=data.index)
data['Adj Close'] = pd.Series([round(val, 2) for val in data['Adj Close']], index=data.index)
stk_data = data[-5:]
col_head = tuple(['Date'] + list(data.columns)) # 'Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume'
arr = stk_data.to_numpy()
plt.figure(linewidth=2, tight_layout={'pad': 0.5}, figsize=(8,6))

ax8 = plt.gca()
ax8.get_xaxis().set_visible(False)
ax8.get_yaxis().set_visible(False)
plt.box(on=None)

# Color map: https://matplotlib.org/3.3.3/tutorials/colors/colormaps.html
ccolors = plt.cm.Blues(np.full(len(col_head), 0.2))

table = plt.table(cellText=arr, loc='center', colLabels=col_head, colColours=ccolors)
table.set_fontsize(16)
table.scale(3, 2.5)
```

## 3D Surface

```python
from mpl_toolkits import mplot3d

fig9 = plt.figure(figsize=(8,5), dpi=100)
ax9 = fig9.add_axes([0.1,0.1, 0.9, 0.9], projection='3d')
z3 = 15 * np.random.random(100)
x3 = np.sin(z3) * np.random.random(100)
y3 = np.cos(z3) * np.random.random(100)
ax9.scatter3D(x3, y3, z3, c=z3, cmap='Blues')
```

```python
def get_z(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

x4 = np.linspace(-6, 6, 30)
y4 = np.linspace(-6, 6, 30)

x4, y4 = np.meshgrid(x4, y4)
z4 = get_z(x4, y4)

fig9 = plt.figure(figsize=(16,8), dpi=100)
ax9 = fig9.add_axes([0.1,0.1, 0.9, 0.9], projection='3d')
ax9.view_init(45, 55)

ax9.set_xlabel('x')
ax9.set_ylabel('y')
ax9.set_zlabel('z')
# ax9.contour3D(x4, y4, z4, 100, cmap='Blues')
ax9.plot_wireframe(x4, y4, z4)
ax9.plot_surface(x4, y4, z4, rstride=1, cstride=1, cmap='Blues', edgecolor='none')
```
